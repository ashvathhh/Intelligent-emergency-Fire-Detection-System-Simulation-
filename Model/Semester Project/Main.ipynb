```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model to fine-tune
model = YOLO('yolov8n.pt')  # yolov8n.pt for a small model; you can choose yolov8s.pt, yolov8m.pt, etc. for larger models

# Train the model with custom data
model.train(
    data='C:/Users/shbha/OneDrive/Documents/Semester Project/dataset.yaml',  # Path to the dataset configuration file
    epochs=50,            # Number of training epochs
    imgsz=640,            # Image size
    batch=16,             # Batch size for training
    workers=2             # Number of data loading workers
)
```
