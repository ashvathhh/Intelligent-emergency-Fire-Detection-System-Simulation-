```python
import cv2
import numpy as np
import tensorflow.lite as tflite

# Load TFLite model
interpreter = tflite.Interpreter(model_path="C:/Users/shbha/OneDrive/Documents/Semester Project/CurrentWeights/best_saved_model/best_float32.tflite")
interpreter.allocate_tensors()

# Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Video capture (0 for webcam or replace with video path)
cap = cv2.VideoCapture("C:/Users/shbha/Downloads/WhatsApp Video 2025-03-11 at 13.22.23_4daaa829.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Resize the frame to match model input size
    input_shape = input_details[0]['shape']
    resized_frame = cv2.resize(frame, (input_shape[1], input_shape[2]))  # Adjust for your model input size
    input_data = np.expand_dims(resized_frame, axis=0).astype(np.float32)  # Adjust dtype if needed

    # Run inference
    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()

    # Get output and process it
    output_data = interpreter.get_tensor(output_details[0]['index'])
    
    # (Modify the following line based on what your model outputs)
    print("Model Output:", output_data)

    # Display video
    cv2.imshow('TFLite Video Stream', frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import torch

def print_shapes(obj, indent=0):
    """Recursively print shapes of tensors in a nested structure."""
    prefix = "  " * indent
    if isinstance(obj, torch.Tensor):
        print(f"{prefix}Tensor shape: {obj.shape}")
    elif isinstance(obj, list):
        print(f"{prefix}List with {len(obj)} elements:")
        for i, item in enumerate(obj):
            print(f"{prefix} Element {i}:")
            print_shapes(item, indent + 1)
    elif isinstance(obj, dict):
        print(f"{prefix}Dict with keys: {list(obj.keys())}")
        for key, value in obj.items():
            print(f"{prefix} Key '{key}':")
            print_shapes(value, indent + 1)
    else:
        print(f"{prefix}{type(obj)}")

# Load your YOLOv8 model (replace "best.pt" with your model file)
model = YOLO("C:/Users/shbha/OneDrive/Documents/Semester Project/CurrentWeights/best.pt")
model.model.eval()  # Set model to evaluation mode

# Display the model architecture (this gives you an idea of the network structure)
print("YOLOv8 Model Architecture:")
print(model.model)

# Define the input image size. YOLOv8 typically uses 640x640 images.
imgsz = 640

# Create a dummy input tensor with shape [1, 3, imgsz, imgsz]
dummy_input = torch.zeros(1, 3, imgsz, imgsz)

# Run the dummy input through the model to get the output structure
with torch.no_grad():
    output = model.model(dummy_input)

print("\nInput shape:", dummy_input.shape)

# Recursively print the structure and shapes of the output(s)
print("\nModel output structure and shapes:")
print_shapes(output)
```
